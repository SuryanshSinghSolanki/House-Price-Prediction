```python
url = "https://www.numbeo.com/quality-of-life/country_result.jsp?country=United+States"
```

```python
html = BeautifulSoup(content.text, 'html.parser')
```

```python
table = html.find('table', {'id' : 't2'})
```

```python
name = table.find('tr').text.split('\n')
```

```python
data = []
for row in table.find_all('tr')[1:6]:
    li = []
    for td in row:
        if td.text != '' and td.text != '\n':
            li.append(td.text)
    data.append(li)
```

```python
df2 = pd.DataFrame(data, columns=name[2:12])
```

```python
df2.head()
```

```python
for col in df2.select_dtypes(include='object').columns:
    if col!='City':
        df2[col] = df2[col].astype(float)
```

```python
df2['zip_code'] = [1001, 1002, 1003, 1004, 1005]
```

```python
df2.shape
```

```python
df2.head()
```

```python
df = df1.merge(df2, on='zip_code' , how='inner')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df = df.drop_duplicates()
df.shape
```

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error , r2_score
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
df.head()
```

```python
df.describe()
```

```python
# IQR
def removeOutlier(col):
    global df
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1
    min_value = Q1 - 3.5*IQR
    max_value = Q3 + 3.5*IQR
    df = df[df[col]>min_value]
    df = df[df[col]<max_value]
```

```python
# Removing outlier
for col in ["CRIM",'ZN']:
    removeOutlier(col)
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
df.CHAS.unique()
```

```python
df.RAD.unique()
```

```python
df.CHAS = df.CHAS.astype(int)
df.RAD = df.RAD.astype(int)
```

```python
df.info()
```

```python
X = df.drop(columns=['MEDV'])
y = df['MEDV']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
cat = X_train.select_dtypes(include='object').columns
```

```python
cat
```

```python
num = X_train.select_dtypes(include=np.number).columns
```

```python
num
```

```python
X_train.head(2)
```

```python
# Numerical frature
num_transformer = Pipeline(steps=[
    ('impute',SimpleImputer(strategy='mean')),
    ('scaled',StandardScaler())
])
```

```python
# Categorical feature
cat_transformer = Pipeline(steps=[
    ('impute',SimpleImputer(strategy='most_frequent')),
    ('encoder',OneHotEncoder())
])
```

```python
preprocessor = ColumnTransformer(transformers= [
    ('numerical',num_transformer,num),
    ('categorical',cat_transformer,cat)
])
```

```python
models = {
    'LinearRegression':LinearRegression(),
    'RandomForestRegressor':RandomForestRegressor(),
    'XGB':XGBRegressor()
}
```

```python
for name,model in models.items():
    pipe = Pipeline(steps=[
        ('preprocessor',preprocessor),
        ('model',model)
    ])
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)
    print(f"Model : {model}")
    print("Mean Squared Error :",round(mean_squared_error(y_pred,y_test),2),'%')
    print("R2_SCORE :",round(r2_score(y_pred,y_test)*100,2),"%")
```
